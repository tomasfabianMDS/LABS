````python
# Escriba su c√≥digo aqu√≠
def clean_data(df):
    df.drop_duplicates(subset=['names'], inplace = True)
    df = df.dropna()
    df = df[(df['math score'] <= 100) & (df['reading score'] <= 100) & (df['writing score'] <= 100)]
    return df
````

- El segundo debe entregar un DataFrame similar al original, pero marcado con una "R" el promedio de notas de los alumnos reprobados.

Para m√°s remate, antes de huir, el practicante les cuenta que este dataframe lamentablemente contiene nuevamente los alumnos de los registros corruptos que ud. y su equipo filtraron en el an√°lisis anterior.


El rector (evidentemente molesto por la situaci√≥n) les ruega incluir estos datos (vaciados en el archivo other_grades.csv) al estudio original(`students_grades.csv`).


Para esto, carge el archivo `other_grades.csv` y busque la forma de unir ambos DataFrames, de tal manera que las columnas de `history score` y `science score` se anexen al final del DataFrame original. **NO LIMPIE LOS DATOS**, si no que explore los distintos tipos de merge para encontrar el mas situable para su situaci√≥n (y as√≠ evitar buscar duplicados).

**To-Do**

- [ ] Unir `df_grades` con `other_grades.csv` usando outer join y explique el resultado.
- [ ] Unir `df_grades` con `other_grades.csv` usando left join y explique el resultado.
- [ ] Unir `df_grades` con `other_grades.csv` usando right join y explique el resultado.

> **Hint**: Puede explicar los resultados del merge a trav√©s de la cantidad de filas resultantes y los valores que estas contienen.

````python

df_other_grades = pd.read_csv('2024040510517433FE38528DF944__other_grades.csv')
print(df_other_grades.isna().sum())
# Unir df_grades con df_other_grades usando outer join
outer_join_result = pd.merge(df_grades, df_other_grades, on='names', how='outer')
display(outer_join_result)
display(outer_join_result.isna().sum())
left_join_result = pd.merge(df_grades, df_other_grades, on='names', how='left')
display(left_join_result)
display(left_join_result.isna().sum())
right_join_result = pd.merge(df_grades, df_other_grades, on='names', how='right')
display(right_join_result)
display(right_join_result.isna().sum())
inner_join_result = pd.merge(df_grades, df_other_grades, on='names', how='inner')
display(inner_join_result)
display(inner_join_result.isna().sum())

````

- Outer join: Devuelve todas las filas de ambos DataFrames, llenando con NaN donde no hay coincidencias entre los DataFrames. En este caso como todas las filas est√°n presentes en ambos DataFrames no hay valores NaN en el resultado del outer join.
  
- Left join:Devuelve todas las filas del DataFrame izquierdo (df_grades) y las filas coincidentes del DataFrame derecho (other_grades.csv). Si no hay coincidencias, las columnas del DataFrame derecho se rellenar√°n con NaN. Aqu√≠, todas las filas de df_grades est√°n presentes, y las columnas science score y history score se llenan con los valores correspondientes de other_grades.csv.
  
- Right join: Es lo opuesto al left join. Devuelve todas las filas del DataFrame derecho (other_grades.csv) y las filas coincidentes del DataFrame izquierdo (df_grades). Si no hay coincidencias, las columnas del DataFrame izquierdo se rellenar√°n con NaN. En este caso, todas las filas de other_grades.csv est√°n presentes, y las columnas de df_grades se llenan con NaN donde no hay coincidencias.
  
- Inner join: Devuelve solo las filas que tienen coincidencias en ambos DataFrames. Es decir, solo devuelve filas para las cuales hay un valor coincidente en ambas tablas. En este caso, solo las filas con nombres que est√°n presentes en ambos DataFrames se incluyen en el resultado.

Utilizar el inner join es la opci√≥n m√°s adecuada para generar el nuevo DataFrame que incluya las notas de historia y ciencias. Esto asegura que solo se incluyan las filas que tienen notas de ambas asignaturas y evita la duplicaci√≥n de datos o la inclusi√≥n de registros corruptos.
