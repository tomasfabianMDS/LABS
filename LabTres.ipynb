````python
# Escriba su código aquí
def clean_data(df):
    df.drop_duplicates(subset=['names'], inplace = True)
    df = df.dropna()
    df = df[(df['math score'] <= 100) & (df['reading score'] <= 100) & (df['writing score'] <= 100)]
    return df
````

### 3.1 Transformación de notas a "Nota Chilena" [0.3 Puntos]

En una de las reuniones con Don Caguayo este les comenta su intención de pasar todas sus notas a la "Escala Chilena". Para esto les facilita la siguiente ecuación:

$$Nota \, Chilena= \% \, de \, logro*0.06 + 1$$

Utilizando la ecuación entregada, cree una función que le permita pasar a la escala Chilena las notas del dataframe y entregue las notas redondeadas al segundo decimal.

**To-Do**

- [ ] Tranforme los valores a la escala chilena.

**Respuesta:**

````python
def nota_chilena(dataframe_in, columna):
    dataframe_in[columna] = dataframe_in[columna].apply(lambda x: round( x * 0.06 + 1, 2))
  #pass
````

Contentos con su desempeño, la Universidad les solicita que obtengan los alumnos quienes han reprobado el año y los que han tenido desempeño de destacado. Cree una función que permita obtener tres Dataframes:
- El primero deberá poseer solamente los alumnos que han reprobado en la prestigiosa universidad.
- El segundo debe entregar un DataFrame similar al original, pero marcado con una "R" el promedio de notas de los alumnos reprobados.

El rector (evidentemente molesto por la situación) les ruega incluir estos datos (vaciados en el archivo other_grades.csv) al estudio original(`students_grades.csv`).


Para esto, carge el archivo `other_grades.csv` y busque la forma de unir ambos DataFrames, de tal manera que las columnas de `history score` y `science score` se anexen al final del DataFrame original. **NO LIMPIE LOS DATOS**, si no que explore los distintos tipos de merge para encontrar el mas situable para su situación (y así evitar buscar duplicados).

> **Hint**: Puede explicar los resultados del merge a través de la cantidad de filas resultantes y los valores que estas contienen.

````python

df_other_grades = pd.read_csv('2024040510517433FE38528DF944__other_grades.csv')
print(df_other_grades.isna().sum())
# Unir df_grades con df_other_grades usando outer join
outer_join_result = pd.merge(df_grades, df_other_grades, on='names', how='outer')
display(outer_join_result)
display(outer_join_result.isna().sum())
left_join_result = pd.merge(df_grades, df_other_grades, on='names', how='left')
display(left_join_result)
display(left_join_result.isna().sum())
right_join_result = pd.merge(df_grades, df_other_grades, on='names', how='right')
display(right_join_result)
display(right_join_result.isna().sum())
inner_join_result = pd.merge(df_grades, df_other_grades, on='names', how='inner')
display(inner_join_result)
display(inner_join_result.isna().sum())

````

- Outer join: Devuelve todas las filas de ambos DataFrames, llenando con NaN donde no hay coincidencias entre los DataFrames. En este caso como todas las filas están presentes en ambos DataFrames no hay valores NaN en el resultado del outer join.
  
- Left join:Devuelve todas las filas del DataFrame izquierdo (df_grades) y las filas coincidentes del DataFrame derecho (other_grades.csv). Si no hay coincidencias, las columnas del DataFrame derecho se rellenarán con NaN. Aquí, todas las filas de df_grades están presentes, y las columnas science score y history score se llenan con los valores correspondientes de other_grades.csv.
  
- Right join: Es lo opuesto al left join. Devuelve todas las filas del DataFrame derecho (other_grades.csv) y las filas coincidentes del DataFrame izquierdo (df_grades). Si no hay coincidencias, las columnas del DataFrame izquierdo se rellenarán con NaN. En este caso, todas las filas de other_grades.csv están presentes, y las columnas de df_grades se llenan con NaN donde no hay coincidencias.
  
- Inner join: Devuelve solo las filas que tienen coincidencias en ambos DataFrames. Es decir, solo devuelve filas para las cuales hay un valor coincidente en ambas tablas. En este caso, solo las filas con nombres que están presentes en ambos DataFrames se incluyen en el resultado.

Utilizar el inner join es la opción más adecuada para generar el nuevo DataFrame que incluya las notas de historia y ciencias. Esto asegura que solo se incluyan las filas que tienen notas de ambas asignaturas y evita la duplicación de datos o la inclusión de registros corruptos.
